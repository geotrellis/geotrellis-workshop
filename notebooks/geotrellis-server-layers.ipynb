### Layers definitions (resources/application.conf)

#### Inidividual Rasters Layers

```conf
    lc8-green = {
        type = "rastersourceconf"
        name = "lc8-green",
        title = "lc8-green"
        source = "gdal+https://s3-us-west-2.amazonaws.com/landsat-pds/L8/003/017/LC80030172015001LGN00/LC80030172015001LGN00_B3.TIF"

        default-style = "red-to-blue"
        styles = [
            {
                name = "red-to-blue"
                title = "Red To Blue"
                type = "colorrampconf"
                colors = ${color-ramps.red-to-blue}
                stops = 64
            }
        ]
    }
```


#### GeoTrellis Layers

```conf
    us-ned = {
        type = "rastersourceconf"
        name = "us-ned"
        title = "US NED"
        source = "gt+s3://azavea-datahub/catalog?layer=us-ned-tms-epsg3857&zoom=14&band_count=1"
        default-style = "elevation-ramp"
        styles = [
            {
                type = "colorrampconf"
                name = "elevation-ramp"
                title = "Elevation Ramp"
                colors = ${color-ramps.elevation}
                stops = 90
            }

        ]
    }
```

#### STAC Layers

```conf
   stac-hillshade-buffered = {
        type = "stacsourceconf"
        with-gdal = true
        name = "hillshade-buffered"
        title = "Hillshade Buffered GeoTrellis Collection."
        layer = "hillshade-buffered"
        asset = "cog"
        asset-limit = 1000
        source = "http://localhost:9090/"
        default-time = true
        default-style = "greyscale"
        styles = [
            {
                name = "greyscale"
                title = "Greyscale"
                type = "colorrampconf"
                colors = ${color-ramps.greyscale}
                stops = 255,
                clamp-with-color = true
            }
        ]
    }
```
